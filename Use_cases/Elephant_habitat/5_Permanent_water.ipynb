## Background

This notebook calculates and stores the first sub-model in the elephant Habitat Suitability Model; the other vegetation sub-model. It has a single permanent water layer.

The notebook executes the following steps:

   1. Define an area of interest and load packages.
   2. Load landcover CGLS data and select the cropland and urban cover layers.
   3. Within the area of interest, calculate proximity to croplands and urban areas quantified as euclidean distance in metres.
   4. Standardise the euclidean distance measure so it ranges between 0 and 1.
   5. Use a weighted linear combination approach to estimate habitat suitability.
   6. Visualise the habitat suitability model.

### Connect to the datacube

Connect to the datacube so we can access DE Africa data.
The `app` parameter is a unique name for the analysis which is based on the notebook file name.

```python
dc = datacube.Datacube(app='Elephant-HSM-pw')
```

## Load CGLS landcover

Spatial parameters must be the same as other notebooks.

```python
lat, lon = -19.26, 17.68

buffer = 3

resolution=(-200, 200) 

measurements='classification'

#convert the lat,lon,buffer into a range
lons = (lon - buffer, lon + buffer)
lats = (lat - buffer, lat + buffer)
```

```python
display_map(x=lons, y=lats)
```

```python
#create reusable datacube query object
query = {
    'x': lons,
    'y': lats,
    'resolution':resolution,
    'time': ('2019'),
    'output_crs': 'epsg:6933',
    'measurements':measurements
}

ds_cgls = dc.load(product='cgls_landcover', **query).squeeze()

ds_cgls
```

## Visualise location of permanent water

```python
plot_lulc(ds_cgls.where(ds_cgls==80)[measurements], product="CGLS", legend=True)
```

```python
cgls_pw = ds_cgls.where(ds_cgls==80)
```

```python
proximity_pw = proximity(cgls_pw.classification, distance_metric = 'EUCLIDEAN')
```

```python
pw_proximity_std = 1 - (proximity_pw/np.max(proximity_pw))
pw_proximity_std = pw_proximity_std.fillna(0)
pw_proximity_std.plot()
```

## Save to outputs

```python
pw_model = pw_proximity_std.drop('time', dim=None)
pw_model.to_netcdf('outputs/pw_sm.nc')
```
